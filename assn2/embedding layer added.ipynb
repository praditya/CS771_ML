```python
import utils
import predict
import time as tm
import numpy as np
from numpy.random import seed
seed(1337)

from keras.models import Sequential, Model
from keras.layers import Activation, Dense, Dropout, Embedding, Flatten
from sklearn.preprocessing import LabelBinarizer
from sklearn.model_selection import train_test_split
from keras.utils import np_utils
from keras.optimizers import SGD

k = 5
dictSize = 225
(X, y) = utils.loadData( "C:\\Users\\asus\\Documents\\GitHub\\ml19-20w\\assn2\\train", dictSize = dictSize )


'''
#tic_DT = tm.perf_counter()
u, c = np.unique(y, return_counts=True)
listOfUniqueValues = zip(u, c)
print('Unique values and their first index :')
# Iterate over the zip object
for elem in listOfUniqueValues:
   print(elem[0], ' at Index : ', elem[1])

'''

# encode class values as integers
#encoder = LabelEncoder()
#encoder.fit(y)
#encoded_Y = encoder.transform(y)
# convert integers to dummy variables (i.e. one hot encoded)
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.33, random_state=42)
Y_train = Y_train-1
dummy_y = np_utils.to_categorical(Y_train)

#dummy_y = y
#print(dummy_y[:,0])
#print(dummy_y.shape)









#let us build a basic model
model = Sequential()
model.add(Embedding(50,7,input_length=225))
model.add(Flatten())
model.add(Dense(128))
#model.add(Dense(128, input_shape=(225,)))
model.add(Dropout(0.1))
model.add(Activation('relu'))
model.add(Dense(128))
model.add(Dropout(0.1))
model.add(Activation('relu'))
model.add(Dense(50))
model.add(Activation('softmax'))
model.summary()

model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
num_epochs = 50
batch_size = 128

history = model.fit(X_train, dummy_y,
                    batch_size=batch_size,
                    epochs=num_epochs,
                    verbose=2, shuffle = True)
```

```python
Y_test = Y_test-1
dummy_yt = np_utils.to_categorical(Y_test)
score, acc = model.evaluate(X_test, dummy_yt,
                       batch_size=batch_size, verbose=2)

print('Test accuracy:', acc)
```
